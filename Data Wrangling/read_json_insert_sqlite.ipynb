```python
# Dependencies
from config import API_KEY
import pandas as pd
import requests
import json
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
```

```python
import sqlite3

conn = sqlite3.connect('JSearchdata.sqlite')
```

```python
conn.execute('''CREATE TABLE IF NOT EXISTS json_data
         (
         id                                             TEXT NOT NULL,
         employer_name                                  TEXT NOT NULL,
         job_publisher                                  TEXT NOT NULL,
         job_employment_type                            TEXT NOT NULL,
         job_title                                      TEXT NOT NULL,
         job_apply_link                                 TEXT NOT NULL,
         job_description                                TEXT NOT NULL,
         job_city                                       TEXT NOT NULL,
         job_state                                      TEXT NOT NULL,   
         job_latitude                                   TEXT NOT NULL,
         job_longitude                                  TEXT NOT NULL,
         job_required_education_bachelors_degree        TEXT NOT NULL);
         ''')

print("Table created successfully")

# conn.close()
```
