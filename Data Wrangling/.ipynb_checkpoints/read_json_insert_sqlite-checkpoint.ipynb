```python
# Dependencies
from config import API_KEY
import pandas as pd
import requests
import json
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
```

```python
import sqlite3

conn = sqlite3.connect('project3.sqlite')
```

```python
conn.execute('''CREATE TABLE IF NOT EXISTS jobs_table
         (
         city                   TEXT NOT NULL,
         employer_name          TEXT NOT NULL,
         employer_company_type  TEXT NOT NULL,
         job_employment_type    TEXT NOT NULL,
         job_title              TEXT NOT NULL,
         job_is_remote          TEXT NOT NULL,
         job_city               TEXT NOT NULL,
         job_latitude           TEXT NOT NULL,
         job_longitude          TEXT NOT NULL);
         ''')
         
print("Table created successfully")

# conn.close()
```

# Read JSON data files and insert into the table

```python
import json
import pprint

cities_list = ['Adelaide', 'Brisbane', 'Canberra', 'Darwin', 'Hobart', 'Melbourne', 'Perth', 'Sydney']

# Read data from a json file 
for city in cities_list:
    file_path = f'../Resources/{city}_data_jsearch.json'

    with open(file_path) as infile:
        jsearch = json.load(infile)  
        print(f"\n{city} has {len(jsearch['data'])}\n")

        for i in range(len(jsearch["data"])):       
#             print(jsearch["data"][i]["employer_name"])
            job_city              = jsearch["data"][i]["job_city"]
            employer_name         = jsearch["data"][i]["employer_name"]        
            employer_company_type = jsearch["data"][i]["employer_company_type"]
            job_employment_type   = jsearch["data"][i]["job_employment_type"]
            job_title             = jsearch["data"][i]["job_title"]
            job_is_remote         = jsearch["data"][i]["job_is_remote"]
            job_latitude          = jsearch["data"][i]["job_latitude"]
            job_longitude         = jsearch["data"][i]["job_longitude"]

            sql_string = f"insert into jobs_table values ('{city}','{employer_name}','{employer_company_type}',\
            '{job_employment_type}','{job_title}','{job_is_remote}','{job_city}','{job_latitude}','{job_longitude}')"

#             print(sql_string)
            conn.execute(sql_string)
    
            conn.commit()
        
        
```
