```python
# Adelaide 

# Save url as variable
url = "https://jsearch.p.rapidapi.com/search"
# Set querystring
querystring = {"query":"Data Analyst in Adelaide, Australia","page":"1","num_pages":"20"}
# Set headers
headers = {
	"X-RapidAPI-Key": API_KEY,
	"X-RapidAPI-Host": "jsearch.p.rapidapi.com"
}
# Make a GET request and save in variable
response = requests.get(url, headers=headers, params=querystring)
Ade_jsearch= response.json()

# Write data to a json file 
with open('../Resources/Adelaide_data_jsearch.json', 'w') as outfile:
	json.dump(Ade_jsearch, outfile, indent=4)
```

### Merge all cities and clean the data 

```python
# Add all the cities dataframes together and format column names
merge_df = pd.concat([Ade_df,Bris_df,Canb_df,Darw_df,Hob_df,Melb_df,Perth_df,Syd_df], axis =0)
# merge_df.columns = merge_df.columns.str.replace("_", " ").str.capitalize()
merge_df
```

```python
# clean data by replacing null with None and remove the \n from the data
jobsearch_df = merge_df.replace({np.NaN: None})
jobsearch_df = merge_df.replace({'\n': ' '}, regex=True)
jobsearch_df.reset_index(drop=True, inplace=True)
jobsearch_df.head(5)
```

```python
# convert dataframe into json final combine file
jobsearch_df.to_json('../Resources/finalcombine.json', indent=3)
```

# Convert JSON file to SQLite database

## Creat an SQLite Database

```python
import sqlite3

# Connect to the SQLite database (it will create a new file if it doesn't exist)
conn = sqlite3.connect('JSearchdata.sqlite')

# Create a cursor object to execute SQL commands
cursor = conn.cursor()

# Create a table to store JSON data
cursor.execute('''
    CREATE TABLE IF NOT EXISTS json_data (
        id INTEGER PRIMARY KEY,
        json_array TEXT
    )
''')

# Commit the changes and close the connection
conn.commit()
conn.close()
```

```python
import sqlite3
import json

# Connect to the SQLite database
conn = sqlite3.connect('JSearchdata.sqlite')
cursor = conn.cursor()

# Read JSON data from a file
with open('../Resources/finalcombine.json', 'r') as file:
    data = json.load(file)

# Insert JSON data into the database
for item in data:
    json_array = json.dumps(item)  # Convert JSON object to a string
    cursor.execute('INSERT INTO json_data (json_array) VALUES (?)', (json_array,))

# Commit the changes and close the connection
conn.commit()
conn.close()
```
